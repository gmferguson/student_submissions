```python
# Further confirming the transformations were useful and adequate
# Looking in more detail at QQ plots and resuduals to see if there need to be further transformations 
# of my variables. Found these to look good.

def diagnostic_plot(x, y, name):
    plt.figure(figsize=(20,5)).suptitle(name)
    
    rgr = LinearRegression()
    rgr.fit(x.reshape(y.size,1),y)
    pred = rgr.predict(x.reshape(y.size,1))
    
    plt.subplot(1, 4, 1)
    plt.scatter(x,y)
    plt.plot(x, pred, color='blue',linewidth=1)
    plt.title("Regression fit")
    plt.xlabel("x")
    plt.ylabel("y")

    plt.subplot(1, 4, 2)
    plt.scatter(pred,y)
    plt.plot(pred, pred, color='blue',linewidth=1)
    plt.title("Regression fit")
    plt.xlabel("predicted y")
    plt.ylabel("observed y")
    
    plt.subplot(1, 4, 3)
    res = y - pred
    plt.scatter(pred, res)
    plt.title("Residual plot")
    plt.xlabel("prediction")
    plt.ylabel("residuals")
    
    plt.subplot(1, 4, 4)
    #Generates a probability plot of sample data against the quantiles of a 
    # specified theoretical distribution 
    stats.probplot(res, dist="norm", plot=plt)
    plt.title("Normal Q-Q plot")
    
    
for col in X_train_transformed.columns:
   diagnostic_plot(X_train_transformed[col].values, y_train_transformed.values,col)
```

Let's see if we should incorporate polynomials...

```python
deg = 4
train_error = np.empty(deg)
test_error = np.empty(deg)

for degree in range(deg):
    est = make_pipeline(PolynomialFeatures(degree, interaction_only=True), StandardScaler(), LinearRegression())
    est.fit(X_train_transformed, y_train_transformed)
    train_error[degree] = mean_squared_error(y_train_transformed, est.predict(X_train_transformed))
    test_error[degree] = mean_squared_error(y_test_transformed, est.predict(X_test_transformed))

# Plot the training and test errors against degree
plt.figure(figsize=(6,6))
plt.semilogy(np.arange(deg), train_error, color='green', label='train')
plt.semilogy(np.arange(deg), test_error, color='red', label='test')
plt.ylabel('log(mean squared error)')
plt.xlabel('degree')
plt.legend(loc='upper left');
```

It looks like MSE for the test set goes up slightly for the 2nd degree, so let's we will stick with 1st degree. Let's now try ridge cross validation to see how a polynomial interaction model performs on average with train and test sets

Now let's try performing ridge and lasso regularization with cross validation. This helps ensure we don't overfit our training data.

```python
est_ridge = make_pipeline(StandardScaler(), RidgeCV(cv=5,alphas=(1e-14,1e-8,1e-2,.1)))

plt.figure(figsize=(6,8))
plt.title('Prediction vs Reality');

prediction_train = est_ridge.fit(X_train,y_train).predict(X_train)
plt.scatter(prediction_train,y_train,c='b',label='train',s=10)

prediction_test = est_ridge.fit(X_test,y_test).predict(X_test)
plt.scatter(prediction_test,y_test,c='r',label='test',s=10)

plt.plot(prediction_train,prediction_train, color='black',linewidth=2)

plt.ylabel('Actual LDL Cholesterol')
plt.xlabel('Predicted LDL Cholesterol')
plt.legend();
# plt.xlim(2,18)
# plt.ylim(2,18)
```
